```python
#removendo linhas que tinham o ano como null
dados.dropna(subset=['Year'], inplace=True)
dados.dropna(subset=['Publisher'], inplace=True)
```

```python
#AGRUPANDO OS ANOS IGUAIS, PARA TER A QUANTIDADE DE LANÇAMENTOS POR ANO
quant_ano = dados.groupby('Year')
quant_ano.groups
```

```python
quant_ano['Name'].count()
```

```python
dados.head(5)
```

```python
fig = quant_ano['Rank'].count().plot.bar(color = 'blue')
fig.set_ylabel('Quantidade')
fig.set_title('Quantidade de lançamentos por ano ', {'fontsize':22})
```

```python
#AGRUPANDO por plataformas.
plataforma = dados.groupby('Platform')
plataforma.groups
```

```python
fig = plataforma['Rank'].count().plot.bar(color = 'blue')
fig.set_ylabel('Quantidade')
fig.set_title('Quantidade de lançamentos por Plataforma ', {'fontsize':22})
```

```python
#AGRUPANDO por Generos.
generos = dados.groupby('Genre')
generos.groups
```

```python
fig = generos['Rank'].count().plot.bar(color = 'red')
fig.set_ylabel('Quantidade')
fig.set_title('Quantidade de jogos por Genero ', {'fontsize':22})
```
